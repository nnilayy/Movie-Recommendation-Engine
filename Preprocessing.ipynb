```python
# Extracting the movies and credits dataset
movies_df=pd.read_csv("tmdb-movie-metadata/tmdb_5000_movies.csv") 
credits_df=pd.read_csv("tmdb-movie-metadata/tmdb_5000_credits.csv") 
movies_df.head()
```

```python
# Merge the two datasets on the 'id' column and drop the 'key_0' column
movies=movies_df.merge(credits_df,on=movies_df['id'])
movies.drop(['key_0'],axis=1,inplace=True)
movies.head()
```

```python
# Drop the columns that are not needed
movies=movies[['id','title_x','genres','keywords','overview','cast','crew','vote_average','vote_count','production_companies']]
movies.head()
```

```python
# Safety check for missing values
movies.isnull().sum()
movies.dropna(inplace=True)
# Check for duplicates
print("Number of Duplicated Rows",movies.duplicated().sum())
```

```python
movies.head()
```

```python
import json
# Extracting Companies
def extract_companies(obj):
  companies=[]
  df=json.loads(obj)
  for i in range(len(df)):
    companies.append(df[i]['name'])
  return companies

#Combining company names as one 
def wordize(obj):
  new_obj = []
  for i in range(len(obj)):
    new_obj.append(obj[i].replace(" ", ""))
  return new_obj

movies['production_companies']=movies['production_companies'].apply(extract_companies)
movies['production_companies']=movies['production_companies'].apply(wordize)
```

```python
# EXTRACTING CAST NAMES
import json
def extract_names(obj):
  names=[]
  df=json.loads(obj)
  for i in range(len(df)):
    names.append(df[i]['name'])
  return names[0:4]

movies['cast']=movies['cast'].apply(extract_names)
```

```python
# EXTRACTING KEYWORDS
def extract_keywords(obj):
  keywords=[]
  df=json.loads(obj)
  for i in range(len(df)):
    keywords.append(df[i]['name'])
  return keywords

movies['keywords']=movies['keywords'].apply(extract_keywords)
```

```python
# EXTRACTING GENRES
def extract_genres(obj):
  genres=[]
  df=json.loads(obj)
  for i in range(len(df)):
    genres.append(df[i]['name'])
  return genres

movies['genres']=movies['genres'].apply(extract_genres)
```

```python
# EXTRACTING DIRECTORS
def extract_director(obj):
  df=json.loads(obj)
  for i in range(len(df)):
    if df[i]['job']=="Director":
      return [df[i]['name']]

movies['crew']=movies['crew'].apply(extract_director)
```

```python
movies.head()
```

```python
# RENAMING COLUMNS
movies.rename(columns = {'title_x':'titles'}, inplace = True)
movies.rename(columns = {'crew':'directors'}, inplace = True)
```

```python
# LISTIFYING OVERVIEW COLUMN
def listify(obj):
  return obj.split()

movies['overview']=movies['overview'].apply(listify)
```

```python
# REMOVING SPACES IN COLUMNS BETWEEN WORDS 
movies['genres']=movies['genres'].apply(wordize)
movies['keywords']=movies['keywords'].apply(wordize)
movies['cast']=movies['cast'].apply(wordize)
```

```python
movies.head()
```

```python
# CREATING THE FINAL TAG COLUMN 
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['directors']+movies['production_companies']
```

```python
# REMOVING OTHER COLUMNS
movies.drop(['genres','keywords','overview','cast','directors','vote_average','vote_count','production_companies'],axis=1,inplace=True)
```

```python
movies.dropna(inplace=True)
movies.reset_index(inplace = True, drop=True)
movies
```

```python
def sentenize(obj):
  sentence= " ".join(obj)
  return sentence

movies['tags']=movies['tags'].apply(sentenize)
```

```python
# Removing Similar Words using Stemming
import nltk 
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()


# Applying Stemming to all the Sentences
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)


# Removing all Similar words using Stemming
movies['tags']=movies['tags'].apply(stem)
```

```python
movies.head()
```

```python
# CONVERTING SENTENCES TO VECTORS USIGN BAG OF WORDS
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=9000,stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
# Checking out the most frequent features/words
cv.get_feature_names_out()
```

```python
# Cosine Similarity finds the cosine distance with every movie
# Which is why all the diagonals are 1, since a movie is same as compared to itself
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
def recommend(movie,num):
  movie_index=movies[movies['titles']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:num]
  
  for i in movies_list:
    print(movies.iloc[i[0]].titles)

# Testing the Recommendation System
recommend('Spider-Man',10)
```
